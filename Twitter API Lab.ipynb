# Lets collect the tweets from Joe Biden's friends

First, we'll create some helper functions to extract certain elements we want from the tweets.

```python
def get_hashtags(text):
    hashtags = []
    for element in text:
        hashtags.append(element['text'])
    return hashtags
```

```python
def get_users_mentioned_username(text):
    username = []
    for element in text:
        username.append(element['screen_name'])
    
    return username


def get_users_mentioned_userid(text):
    userid = []
    for element in text:
        userid.append(element['id_str'])
    
    return userid
```

```python
def get_media_url(text):
    url = []
    try: 
        for element in text:
            url.append(element['media_url'])
    except:
        pass

    return url
```

Now, lets create a function that gets the tweets from a given user's timeline, puts them into a pandas dataframe, and then does a bit of processing using the helper functions above

```python
def get_timeline_tweets(id):
    results=[]

    timeline = tw.Cursor(api.user_timeline, user_id = id, tweet_mode="extended",count=200).items()
    for status in timeline:
        data = (
        status.id,    
        status.user.id,
        status.user.name,    
        status.user.screen_name,
        status.user.location,
            status.user.description,
            status.user.followers_count,
            status.user.friends_count,
            status.user.statuses_count,
            status.user.created_at,
            status.user.verified,
            status.full_text,
            status.created_at,
            status.source,
            status.in_reply_to_user_id,
            status.in_reply_to_status_id,
            #status.quote_count,
            #status.reply_count,
            status.favorite_count,
            status.retweet_count,
            status.entities)
        results.append(data)


    cols = "tweetid userid user_display_name user_screen_name user_reported_location user_profile_description followers_count following_count num_tweets account_creation_date verified tweet_text tweet_time tweet_client_name in_reply_to_userid in_reply_to_tweetid like_count retweet_count entities".split()
    df = pd.DataFrame(results, columns=cols)
    expanded = df['entities'].apply(lambda row: pd.Series(row))
    df = pd.concat([df, expanded], axis=1)
    df['hashtags'] = df['hashtags'].apply(lambda x: get_hashtags(x))
    df['username_mentioned'] =df['user_mentions'].apply(lambda x: get_users_mentioned_username(x)) 
    df['userid_mentioned'] = df['user_mentions'].apply(lambda x: get_users_mentioned_userid(x))
    df['media_url'] = df['media'].apply(lambda x: get_media_url(x))
    df = df.drop(columns=['entities','symbols','media','users_mentioned'])
    return df
    
```

Now, lets create a function that iterates through a list of users, calls the above function to get their tweets, and concatenates all user dataframes into one big dataframe

```python
def get_multi_users_timeline_tweets(users):
    cols = "tweetid userid user_display_name user_screen_name user_reported_location user_profile_description followers_count following_count num_tweets account_creation_date verified tweet_text tweet_time tweet_client_name in_reply_to_userid in_reply_to_tweetid like_count retweet_count entities".split()
    df = pd.DataFrame(columns=cols)
    dfs = []
    for user in users:
        dfs.append(get_timeline_tweets(user))
        
    tweets = pd.concat([df.squeeze() for df in dfs], ignore_index=True)
    return tweets
```

Of the tweets we collected, what were the most common hashtags?

```python
from collections import Counter
def find_top_hashtags(df, topn):
    all_tweets = df.shape[0]
    hashtags = df.loc[df['hashtags'].notnull()]
    hashtags = hashtags.loc[hashtags['hashtags']!='[]']
    hashtags['hashtags'] =hashtags['hashtags'].map(lambda  x: str(x))
    hashtags['hashtags'] =hashtags['hashtags'].map(lambda  x: x.lower())
    hashtags['hashtags'] =hashtags['hashtags'].map(lambda x: re.sub('[^a-zA-Z]', '', x))
    hashtags['hashtags'] =hashtags['hashtags'].map(lambda x: re.sub('\W+', '', x))
    #hashtags['hashtags'] =hashtags['hashtags'].map(lambda x: re.sub(r'[0-9]+', '', x))
    hashtags['hashtags'] = hashtags['hashtags'].map(lambda x: x.split())
    hashtags_list = hashtags['hashtags'].tolist()
    hashtags_flat = [item for sublist in hashtags_list for item in sublist]
    counter = Counter(hashtags_flat).most_common(topn)
    res = []
    column_name = 'Hashtag: ' 
    for c in counter:
        
        res.append({column_name: c[0], 'Count':  c[1],'Percentage (%)': c[1]/len(hashtags_flat)*100})
    return res
```

```python
!pip install stop-words
from stop_words import get_stop_words

stop_words = get_stop_words('en')
import nltk
nltk.download('stopwords')
WPT = nltk.WordPunctTokenizer()



```

```python
def clean_text(text):
    text = text.lower()
    text = text.strip()
    
    text = re.sub(r"(?:\@|https?\://)\S+", "", text)
    text = re.sub('(?P<url>https?://[^\s]+)', ' ', text)
    text = re.sub('\W+', ' ', text)
    text = re.sub('rt', ' ', text)
    text = re.sub(" \d+", " ", text)
    pattern = r"[{}]".format(",.;") 
    text = re.sub(pattern, "", text) 
    WPT = nltk.WordPunctTokenizer()
    tokens = WPT.tokenize(text)
    # EN: Filter out the stop-words 
    tokens = [w for w in tokens if len(w) > 2] 
    filtered_tokens = [token for token in tokens if token not in stop_words]
    # EN: Reconstruct the document
    text = ' '.join(filtered_tokens)
    
    return text
```

```python
from collections import Counter
from nltk import ngrams
import matplotlib.pyplot as plt

import itertools
import collections


import nltk
from nltk.corpus import stopwords
```

```python
def most_common(df):
    df['tweet_text_cleaned'] = df['tweet_text'].apply(lambda x: clean_text(x))
    tweets_list = df['tweet_text_cleaned'].to_list()
    words_in_tweet = [tweet.split() for tweet in tweets_list]
    # List of all words across tweets
    all_words_no_urls = list(itertools.chain(*words_in_tweet))

    # Create counter
    counts_no_urls = collections.Counter(all_words_no_urls)
    
    from collections import Counter
    from nltk import ngrams

    flat_list = [item for sublist in words_in_tweet for item in sublist]
    ngram_counts2 = Counter(ngrams(flat_list, 2))
    ngram_counts3 = Counter(ngrams(flat_list, 3))
    
    return  counts_no_urls,  ngram_counts2,  ngram_counts3
```
